```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import data
data1 = pd.read_csv('csv_data/minimum_runtime.csv') 
data2 = pd.read_csv('csv_data/network_runtime.csv')
data3 = pd.read_csv('csv_data/mincostflow_runtime.csv')
```

```python
plt.figure(figsize=(10, 6))  # Kích thước của biểu đồ

# Plot dữ liệu
plt.plot(data1['run_time'], marker='o', linestyle='-')
plt.plot(data2['run_time'], marker='o', linestyle='-')
plt.plot(data3['run_time'], marker='o', linestyle='-')

# Đặt tên cho đường biểu đồ
plt.legend(['Successive Shortest Path runtime', 'Network Simplex Method runtime', 'Minimum Cost Flow runtime'], loc='upper left') 

# Đặt tên cho trục x và y
plt.xlabel('Kích thước grid')  # Thay 'Số thứ tự' bằng tên phù hợp với dữ liệu của bạn
plt.ylabel('Thời gian chạy')  # Thay 'Thời gian chạy' bằng tên phù hợp với dữ liệu của bạn

# Đặt tiêu đề cho biểu đồ
plt.title('Biểu đồ thời gian chạy')  # Thay 'Biểu đồ thời gian chạy' bằng tiêu đề phù hợp với dữ liệu của bạn

# Hiển thị biểu đồ
plt.grid(True)  # Hiển thị lưới
plt.tight_layout()  # Tạo độ rộng cho biểu đồ
plt.show()
```
