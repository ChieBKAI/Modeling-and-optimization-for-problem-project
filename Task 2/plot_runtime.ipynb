```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import data
data1 = pd.read_csv('csv_data/minimum_runtime.csv') 
data2 = pd.read_csv('csv_data/network_runtime.csv')
data3 = pd.read_csv('csv_data/mincostflow_runtime.csv')
```

```python
plt.figure(figsize=(10, 6))  # Kch thc ca biu 

# Plot d liu
plt.plot(data1['run_time'], marker='o', linestyle='-')
plt.plot(data2['run_time'], marker='o', linestyle='-')
plt.plot(data3['run_time'], marker='o', linestyle='-')

# t tn cho ng biu 
plt.legend(['Successive Shortest Path runtime', 'Network Simplex Method runtime', 'Minimum Cost Flow runtime'], loc='upper left') 

# t tn cho trc x v y
plt.xlabel('Kch thc grid')  # Thay 'S th t' bng tn ph hp vi d liu ca bn
plt.ylabel('Thi gian chy')  # Thay 'Thi gian chy' bng tn ph hp vi d liu ca bn

# t tiu  cho biu 
plt.title('Biu  thi gian chy')  # Thay 'Biu  thi gian chy' bng tiu  ph hp vi d liu ca bn

# Hin th biu 
plt.grid(True)  # Hin th li
plt.tight_layout()  # To  rng cho biu 
plt.show()
```
